```python
import base64 
```

```python
def get_token(client_id, client_secret):
    encoded = base64.b64encode(bytes(client_id+':'+client_secret, 'utf-8'))
    params = {'grant_type':'client_credentials'}
    header={'Authorization': 'Basic ' + str(encoded, 'utf-8')}
    r = requests.post('https://accounts.spotify.com/api/token', headers=header, data=params)
    if r.status_code != 200:
        print('Error en la request.', r.json())
        return None
    print('Token válido por {} segundos.'.format(r.json()['expires_in']))
    return r.json()['access_token']
```

# Función para obtener discografía

```python
def obtener_discografia(artist_id, token, return_name = False, page_limit =50, country = None):
    url = f'https://api.spotify.com/v1/artists/{artist_id}/albums'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit, 
              'offset': 0,
              'country': country}
    lista = []
    r = requests.get(url, params=params, headers=header)
    
    if r.status_code != 200:
        print('Error en request.', r.json())
        return None
    
    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        lista += [item['id'] for item in r.json()['items']]
        
    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header) # El resto de los parámetros están dentro de la URL
        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [item['id'] for item in r.json()['items']]
    
    return lista
```

```python
obtener_discografia(id_thexx, token, return_name=True)
```

# Función para obtener los tracks del album

```python
def obtener_tracks(album_id, token, return_name=False, page_limit=50, country=None):
    url=f'https://api.spotify.com/v1/albums/{album_id}/tracks'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit, 
              'offset': 0,
              'market': country}
    lista = []
    r = requests.get(url, params=params, headers=header)
    
    if r.status_code != 200:
        print('Error en request.', r.json())
        return None
    
    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        lista += [item['id'] for item in r.json()['items']]
        
    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header) # El resto de los parámetros están dentro de la URL
        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [item['id'] for item in r.json()['items']]
    
    return lista
```

```python
obtener_tracks('1hS6yQqEmHAsZeDxu0wtWK', token, return_name=True, country='MX')
```

Ahora podemos usar ambas funciones en combinación para obtener la lista de temas disponibles en México de The XX, cada una con referencia a su album

```python
for album in obtener_discografia(id_thexx, token, return_name=True, country='MX'):
    print(album[1])
    for track in obtener_tracks(album[0], token, return_name=True, country='MX'):
        print('\t',track[1])   
```
